        - average injuries per week across four seasons

```python
# ------- type the answer below this line ------- #
print("# ---------- Answer # 1 -------------------")
all_seasons = ['summer', 'fall', 'winter', 'spring']

### Get data: rating
rsm = stat_summer.loc['mean','ratings']
rf = stat_fall.loc['mean','ratings']
rw = stat_winter.loc['mean','ratings']
rsg = stat_spring.loc['mean','ratings']
# Create lists
avg_cr = [rsm, rf, rw, rsg] # <-- data is in same sequence or order as the list all_seasons


### Get data: injuries
ism = stat_summer.loc['mean','injuries']
i_f = stat_fall.loc['mean','injuries']
iw = stat_winter.loc['mean','injuries']
isg = stat_spring.loc['mean','injuries']
# Create lists
avg_ir = [ism, i_f, iw, isg]



### Get data: visitors
vsm = stat_summer.loc['mean','visitors']
vf = stat_fall.loc['mean','visitors']
vw = stat_winter.loc['mean','visitors']
vsg = stat_spring.loc['mean','visitors']
# Create lists
avg_vr = [vsm, vf, vw, vsg]

# Create plots
fig = plt.figure(figsize=(10,4))

plt.subplot(1,3,1)
plt.bar(all_seasons, avg_cr)
plt.xlabel("Season")
plt.ylabel("Average rating")
plt.title("Customer Rating.")


plt.subplot(1,3,2)
plt.bar(all_seasons, avg_ir)
plt.xlabel("Season")
plt.ylabel("Average rating")
plt.title("Injuries.")


plt.subplot(1,3,3)
plt.bar(all_seasons, avg_vr)
plt.xlabel("Season")
plt.ylabel("Average rating")
plt.title("Visitors.")

plt.suptitle("Statistics for Wonderland Resort", fontsize=17)
plt.tight_layout(rect=[0, 0.13, 1, 0.95])
plt.show()
```

### Question 4:

Based on your analysis, the resort team would like to update their dataset with new activities. Their plan is to use it later to generate predictions for the coming year.

#### Requests:
Create a copy of 'dfs' and save as 'dfs_udpated':
    - In 'dfs_updated' remove the activities by season that has counts of visitors <250 and ratings <3.0
        
    - Add new activities from the Series listed below to above 'dfs_updated'
        [Hint: Before appending to dfs, make sure to add appropriate indices to each Series ]
        [Hint: Use .reset_index() to convert index to columns. Rename the column for seasons as 'seasons']

    - Finally, sort the 'dfs_updated' by 'seasons'. Then, convert the 'seasons' column back to index
    - And find the shape of this DataFrame
    
    
Print all answers to output.

[Note: Create new variables of your choice as needed. Create new cells as needed]

```python
# New data:              season, visitors, week, activity,   injuries, rating (<-- expected)
activity_1 = pd.Series(['summer', 500,      0,  'waterslide', 0,         4])
activity_2 = pd.Series(['winter', 400,      0,  'snowbuilding', 0,         4])
activity_3 = pd.Series(['fall', 300,      0,  'fruitsplash', 0,         5])
activity_4 = pd.Series(['spring', 400,      0,  'travelabroad', 0,         5])
```

```python
# ------- type the answer below this line ------- #


# Add indices to the Series that are same as column names in dfs_updated
a_index = ['seasons', 'visitors', 'week', 'activities',   'injuries', 'ratings']
activity_1 = pd.Series(['summer', 500,      0,  'waterslide', 0,         4], index=a_index)
activity_2 = pd.Series(['winter', 400,      0,  'snowbuilding', 0,         4], index=a_index)
activity_3 = pd.Series(['fall', 300,      0,  'fruitsplash', 0,         5], index=a_index)
activity_4 = pd.Series(['spring', 400,      0,  'travelabroad', 0,         5], index=a_index)

# Create a list of Series to use later for append
a_list = [activity_1,activity_2,activity_3,activity_4]

# Create a copy
dfs_updated = dfs.copy()

# Remove rows that are: visitors <250 and ratings <3.0
print('Before removing rows: ', dfs_updated.shape)
dfs_updated = dfs_updated[~((dfs_updated['visitors']<250)&(dfs_updated['ratings']<3.0))]
print('After removing rows: ', dfs_updated.shape)
print("\n")

# Reset index to convert index to columns
dfs_updated = dfs_updated.reset_index()

# Rename the columns converted from index
dfs_updated = dfs_updated.rename(columns={'level_0':'seasons'})

# Drop column not needed
dfs_updated = dfs_updated.drop('level_1', axis=1)

# Append the Series
dfs_updated = dfs_updated.append(a_list, ignore_index=True)

# Convert the 'seasons' column back to Index
dfs_updated = dfs_updated.set_index('seasons')

# Sort
dfs_updated = dfs_updated.sort_values(by='seasons')

print("Shape of dfs_updated: ", dfs_updated.shape)
dfs_updated
```

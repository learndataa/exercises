### Question 1:

Imagine working at a dinosaur museum. The team there has data on jurassic era that needs to be put in a dataframe format. And you step-in with your Pandas skills to create tuples, pivot, stack/unstack dataframes, create multi-column/row indicies to help. 
Millions of years ago the Jurassic period spanned three epochs. The data is listed in the cell below. 

#### Request:
    - (i) Create a dataframe (df) that has multi level row index
    - (ii) Convert df so that epochs are column names
    - (iii) Convert df so that the ages are column names

    Hint: First create MultiIndex using tuples or arrays, for-loops. 
    Note: There are multiple ways to answer the question

```python
#########
### Data
#########

# Epochs
epochs = ['upper/late jurassic', 'middle jurassic', 'lower/early jurassic']

# Ages
upper_jurassic = ['tithonian', 'kimmeridgian', 'oxfordian']
middle_jurassic = ['callovian', 'bathonian', 'bajocian', 'aalenian']
lower_jurassic = ['toarcian', 'pliensbachian', 'sinemurian', 'hettangian']

# Million years ago (mya)
mya = np.array([
    # upper jurassic
    np.random.randint(145, 152, 20), # tithonian
    np.random.randint(152, 157, 20), # kimmeridgian
    np.random.randint(157, 163, 20), # oxfordian
    
    np.random.randint(163, 166, 20), # callovian
    np.random.randint(166, 168, 20), # bathonian
    np.random.randint(168, 170, 20), # bajocian
    np.random.randint(170, 174, 20), # aalenian
    
    np.random.randint(174, 182, 20), # toarcian
    np.random.randint(182, 190, 20), # pliensbachian
    np.random.randint(190, 199, 20), # sinemurian
    np.random.randint(199, 201, 20)  # hettangian
])
```

```python
# ------- type the answer below this line ------- #
print("# ---------- Answer # 1(i) -------------------\n\n")

# Create index tuples
idx = []
ages = [upper_jurassic, middle_jurassic, lower_jurassic]
for epoch in epochs:
    for age in ages:
        for i in range(len(age)):
            idx.append((epoch, age[i]))
        break;
    ages.pop(0)

print('Index tuples:', idx, '\n')


# Create MuliIndex
idx_jurassic = pd.MultiIndex.from_tuples(idx, names=["epoch", "age"])
print(idx_jurassic, '\n')


# Create dataframe with MultiIndex rows
df = pd.DataFrame(mya, index=idx_jurassic) #<- Transpose to flip rows to columns
df
```

```python
print("# ---------- Answer # 1(ii) -------------------\n\n")
print("Unstack level=0 for epoch to column names")
df.unstack(level=0)
```

```python
print("# ---------- Answer # 1(ii) -------------------\n\n")
print("Unstack level=0 for ages to column names")
df.unstack(level=1)
```

### Question 2:

```python
# ------- type the answer below this line ------- #
print("# ---------- Answer # 2(i) -------------------\n\n")

# Transorm the df from Q#1 to convert row indices to column names
df2 = df.copy() 

df2 = df2.T
df2.head(2)

```

```python
print("# ---------- Answer # 2(ii) -------------------\n\n")

# Stack 
x = df2.stack(level=0)
x.head(3)
```

```python
print("# ---------- Answer # 2(iii) -------------------\n\n")

# Stack
y = df2.stack(level=1)
y.head(3)
```

```python
print("# ---------- Answer # 2(iii) -------------------\n\n")

# Stack 
z = df2.stack(level=[0,1])
z.head(5)
```

```python
### Set random seed
np.random.seed(1)

###################################
# * * * Set image dimensions * * * 
###################################
img_width = 5
img_height = 5
img_size = img_width*img_height
###################################

# DataFrame with image data
df = pd.DataFrame({
    'noise':np.random.randint(0,255,img_size),
    'image':np.linspace(0,255,img_size),
    'noisy_image':np.random.randint(0,255,img_size)+np.linspace(0,255,img_size),
})
print(df.shape)
```

```python
###################################
### ----------- Example -----------
###################################
np.random.seed(1) # <-- Random Seed to generate same set of random values when using 'np.random'
# Image noise
x=np.array(df['noise']).reshape(img_width,img_height)
plt.imshow(x)
```

```python
np.random.seed(1)

# ------- type the answer below this line ------- #

print("# Answer -----------------------------------")

# Noise
noise = np.array(df['noise']).reshape(img_width,img_height)

# Image
img = np.array(df['image']).reshape(img_width,img_height)

# Noisy image
noisy_image = np.array(df['noisy_image']).reshape(img_width,img_height)

# Subtract Noise from Noisy_image
clear_img = img - noisy_image


# Create subplots for above four images: 1 row and 4 columns
###################
### Create images
###################
fig = plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
# Image noise
plt.xlabel("Image size: "+ str(img.shape)) 
plt.title('Image')
plt.imshow(img)

plt.subplot(1,3,2)
# Clear image without noise
plt.xlabel("Image size: "+ str(noise.shape))
plt.title('Noise')
plt.imshow(noise)

plt.subplot(1,3,3)
# Actual image with noise
plt.xlabel("Image size: "+ str(noisy_image.shape))
plt.title('Image + noise')
plt.imshow(noisy_image)

plt.tight_layout()
plt.show()

```

```python
print("# Answer -2 -----------------------------------")

# Create a noisy image
noisy_image_2 = noise + img

# Plot the noisy image
plt.xlabel("Image size: "+ str(noisy_image_2.shape))
plt.title('Image with noise')
plt.imshow(noisy_image_2)
```

```python
print("# Answer -3 -----------------------------------")

# Create a noisy image
clear_img = noisy_image_2 - noise

# Plot the noisy image
plt.xlabel("Image size: "+ str(clear_img.shape))
plt.title('Clearer image after noise reduction')
plt.imshow(clear_img)
```
